```python
'''
El objetivo de este código es realizar webscraping 
a la página "http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx" 
que contiene archivos en formato xls.

Se realizó un breve estudio del DOM (Document Object Model) con la finalidad de extraer las URLs esenciales.

Requisitos:

Es preciso tener instalado Python version 3.7.0 y las librerias requests, bs4 y urllib.request.

Para instalar las librerias de Python en Windows es preciso instalar pip des la consola de Windows mediante:

    python get-pip.py

Luego escribir:

    pip install bs4
    urllib.request

Esta es una versión en desarrollo. Se están afinando los parámetros para sincronizar la solicitud realizada al servidor 
con su respuesta de modo que este no interprete el continuo número de solicitudes con un ataque.   


Instrucciones

Para ejecutarlo localmente es preciso cambiar la variable OUTPUT_DIR a la dirección deseada
'''

import os
import re
import time
import requests
from time import sleep
from random import randint
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import urlopen, urlretrieve, Request

URL = ["http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FAGC&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FBolsa&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FContratos&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FDesviaciones&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?Paged=TRUE&p_SortBehavior=0&p_FileLeafRef=Precio_Oferta_%28%24kWh%29_2010%2exlsx&p_ID=2385&RootFolder=%2ftransacciones%2fHistorico%20Transacciones%2fPrecios&PageFirstRow=101&&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",        
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FPrecios&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FReconciliaciones&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FRestricciones&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}",
"http://informacioninteligente10.xm.com.co/transacciones/Paginas/HistoricoTransacciones.aspx?RootFolder=%2Ftransacciones%2FHistorico%20Transacciones%2FServicios&FolderCTID=0x0120000E28F31AB3520A4C8D8CB6881FFA0AFE&View={75B6225D-9047-44EC-8EC5-DEA7F7AA139C}"
]

OUTPUT_DIR = r'C:\Users\Usuario\Desktop\jairo\ArchivosExcel'  # ruta de salida
t = "http://informacioninteligente10.xm.com.co/transacciones/_layouts/xlviewer.aspx?id="

for url in URL:
    try:
        html_page = requests.get(url).text #ojo: meterle un delay para decirle al requests 
        #que deje de esperar una respuesta despues de un numero de segundos (en este caso parametro timeout entre 3 y 8)
        #,timeout=randint(3,8). USANDO CON POST
        soup = BeautifulSoup(html_page, "html.parser")
        for link in soup.find_all('a'):
            #time.sleep(randint(1,3))
            href = link.get('href')
            if not href  is None:
                href = link.get('href') # href es un string
                if not any(href.endswith(x) for x in ['.xls','.xlsx','.csv']):
                    continue
                href = re.sub(r"\s+", "", href, flags=re.UNICODE) #Elimina los espacios en blanco de href
                href = t + href
                filename = os.path.join(OUTPUT_DIR, href.rsplit('/', 1)[-1])

                #href = href.replace('http://','http://') para reemplazar http por https

                print("Descargando %s a %s..." % (href, filename))

                urlretrieve(href, filename)

                print("Listo.")
                #time.sleep(randint(10,20))
    except:
        print("-"*30)
        print("Error 500")
        print(href)
        print("-"*30)
        
    print("Se han descargado todos los archivos de " + url)    
print ("Descarga completa") 
```
